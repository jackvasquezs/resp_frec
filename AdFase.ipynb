### 2. Control de adelanto de fase (viga y bola)

El sistema de viga y bola se muestra en la siguiente figura.

![ballbeam](img/bbfig.png)

La funci칩n de transferencia para este sistema tiene la forma:

$$G_{BB}(s)=\frac{22.5}{28.3s^2+16s+1}$$

El controlador de adelanto de fase tiene la siguiente funci칩n de transferencia:

$$G_c(s)=\frac{1+aTs}{1+Ts}$$

El dise침o de este controldor se realiza en el dominio de la frecuencia. Para calcular sus valores es necesario obtener las gr치ficas de Bode.

```python
sys = TransferFunction(22.5, [28.3,16,1])
sys
```

```python
mag, phase, omega = bode(sys)
```

En la gr치fica se puede observar que el margen de fase, el cual se obtiene cuando la curva de ganancia cruza por cero es cercano a los 40 grados. En la siguiente gr치fica se muestra la respuesta en el tiempo del sistema retroalimentado.

```python
sysr=co.feedback(sys,1,-1)
tr,yr=co.step_response(sysr)
plt.plot(tr,yr)
plt.grid()
plt.ylabel('Velocidad ($\omega$)')
plt.xlabel('Tiempo (t)')
#plt.ylim(0,1.20)
#plt.xlim(0,3)
plt.show()
```

Al aumentar el margen de fase se mejora la respuesta en el tiempo del sistema. Para lograr un mayor margen de fase se debe atenuar la ganancia. De tal manera que se busca en la gr치fica la frecuencia a la cual se obtienen un mayor margen de fase. 

```python
mag, phase, omega = bode(1.25*sys)
```

Se desea un margen de fase $60^o$ es decir hay que agregar $30^o$, de tal manera que por seguridad se agregan $\phi_m=45^o$, con la siguiente ecuaci칩n se calcula el valor de $a$:

$$a=\frac{1+sen \phi_m}{1-sen\phi_m}$$

```python
rad=55*math.pi/180
a=(1+math.sin(rad)/(1-math.sin(rad)))
a
```

El controlador aumentar치 la fase, pero tambi칠n la ganancia, para lograr el incremento en la fase, es necesario buscar la frecuencia donde la magnitud del sistema no compensado es igual a $-10log10(a)$

```python
omega = bode(1.25*sys)
```

```python
-10*math.log10(a)
```

La ganancia se obtiene a la frecuencia de 0.2 Hz, con este dato y la siguiente ecuaci칩n se calcula el valor de T. 

$$T=\frac{1}{\pi f_m\sqrt {a}}$$

```python
fm=0.25
T=1/(math.pi*fm*math.sqrt(a))
T
```

Ahora se obtiene la respuesta del sistema con el controlador.

```python
contr=co.tf([a*T,1],[T, 1])
contr
```

```python
sysf=contr*sys
sysfr=co.feedback(sysf,1,-1)
trf,yrf=co.step_response(sysfr)

plt.plot(tr,yr)
plt.plot(trf,yrf)
#plt.plot(t2, [1 for i in t2])
plt.grid()
plt.ylabel('Posici칩n ')
plt.xlabel('Tiempo (t)')
#plt.ylim(0,1.20)
#plt.xlim(0,150)
plt.show()
```
