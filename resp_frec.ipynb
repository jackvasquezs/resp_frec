### 2. Respuesta en frecuencia

El término respuesta en frecuencia, muestra la respuesta en estado estacionario de un sistema a una entrada senoidal. En el análisis de la frecuencia, los métodos no están restringidos a funciones de segundo orden. Es posible diseñar controladores para sistemas de mayor orden. Otra caracterísitica a favor es que mediante este análisis es posible determinar la sensibilidad al ruido y la variación de parámetros. Los métodos comúnmente empleados son: *Bode* y *Nyquist*.

##### 2.1.1 Respuesta en frecuencia en lazo cerrado

La función de transferencia del sistema retroalimentado en términos de la frecuencia, cambiando $s$ por $j\omega$, queda expresada de la siguiente manera,  :

\begin{equation}
M(j\omega)=\frac{Y(j\omega)}{R(j\omega)}=\frac{G(j\omega)}{1+G(j\omega)H(j\omega)}
\end{equation}

En términos de la magnitud: 
    
\begin{equation}
|M(j\omega)|=  \left| \frac{Y(j\omega)}{R(j\omega)} \right|=  \left| \frac{G(j\omega)}{1+G(j\omega)H(j\omega)}\right|
\end{equation}

y el ángulo:

\begin{equation}
\angle M(j\omega)=\angle G(j\omega)-\angle(1+G(j\omega)H(j\omega))
\end{equation}
 
Para un sistema prototipo de segundo orden con retroalimentacion unitaria, se tiene que:

\begin{equation}
 M(s)=\frac{M(s)}{R(s)}=\frac{w_n^2}{s^2+2\zeta \omega_ns+\omega_n^2}
\end{equation}

La respuesta en frecuencia se obtiene al sustituir  $s$ por $j\omega$, y resulta en:

\begin{equation}
M(j\omega)=Y(j\omega)/R(j\omega)=\frac{1}{1-(\frac{\omega}{\omega_n})^2+j2\zeta\frac{\omega}{\omega_n}}
\end{equation}

haciendo $u=\frac{\omega}{\omega_n}$

\begin{equation}
M(ju)=\frac{1}{1-u^2+j2u\zeta}
\end{equation}

De aquí que la magnitud y fase estén determinadas por:

\begin{equation}
\left| M(ju) \right| =\frac{1}{\sqrt{(1-u^2)^2+(2u\zeta)^2}}
\end{equation}

\begin{equation}
\angle {M(ju)}=-tan^{-1} \left\{ \frac{2\zeta u}{1-u^2} \right \}
\end{equation}

Si se grafica la respuesta en magnitud del sistema de segundo contra el valor de *u* variable, es posible obtener las siguientes especificaciones:

* Pico de resonancia $M_r$, es valor máximo de $\left| {M(ju)} \right| $. 
* Frecuencia de resonancia $\omega_r$, es donde ocurre el pico de resonancia $M_r$.
* Ancho de banda **BW** es la frecuencia en la cual $\left| {M(ju)} \right|$ cae al 70.7\% de su valor en la frecuencia cero.

#### 3.1.4 Ancho de banda en función de $\zeta$
El ancho de banda es directamente proporcional a la frecuencia natural amortiguada $\omega_n$. Cuando $\omega_n$ se incrementa el sistema responde más rápido. También el **BW** disminuye con el incremento de $\zeta$. El ancho de banda está dado por la siguiente ecuación:

\begin{equation}
BW=w_n \left\{ (1-2\zeta^2)+\sqrt{4\zeta^4-4\zeta^2+2} \right\}^{1/2}
\end{equation}

### 3.2 Efectos de la adición de polos y ceros en la trayectoria directa

### 3.3 Ecuación característica
La función de transferencia para un sistema con retroalimentación unitaria SISO es:

\begin{equation}
M(s)=\frac{G(s)}{1+G(s)H(s)}
\end{equation}

Donde la ecuación característica puede escribirse como:

\begin{equation}
1+G(s)H(s)=1+L(s)=0
\end{equation}

L(s) se conoce como la *función de transferencia de lazo*. Para retroalimentación unitaria H(s)=1 y L(s)=G(s).
 
**Ventajas:**
	
* El cruce de ganancia ($\omega_p$), el cruce de fase ($\omega_g$), el margen de ganancia (GM) y el margen de fase (PM) se determinan fácilmente en las trazas de Bode.  

* Para propósitos de diseño, los efectos de añadir controladores y sus parámetros se visualizan con mayor facilidad sobre las gráficas de Bode.

* Las gráficas de Bode se obtienen de la función de transferencia de lazo L(s).


### 3.4 Análisis mediante gráficas de Bode

Después de obtener la gráfica de Bode es posible considerar los siguientes puntos.

#### 3.4.1 Propiedades de la gráfica de Bode
	
* El margen de ganancia es positivo y el sistema es estable si la magnitud de L(j$\omega$) al cruce de fase es negativo en dB. Esto es, el margen de ganancia se mide abajo del eje 0 dB. 
* El margen de ganancia se define como $GM=-20 log_{10} |L(j \omega_p)|$
* El margen de fase es positivo y el sistema es estable si la fase de L(j$\omega$) es mayor que -180$^o$ en el cruce de ganancia. Esto es, el margen de fase se mide arriba del eje -180$^o$. 
* El margen de fase se define como $PM =\angle |L(j \omega_p)| -180$

<img src="mfmg.png" width="800px">

  

	
#### 3.4.2 Gráficas de Bode con Scilab
Para obtener las gráficas de Bode se emplean los siguientes comandos:
	
\begin{verbatim}
bode(sys)
show_margin(sys)
g_margin(sys)
p_margin(sys)
\end{verbatim}
	
**Ejemplo:** Para la función de transferencia de lazo $L(s)=G(s)H(s)=\frac{2500}{s(s+5)(s+50)}$, se obtienen las gráficas de Bode.

\begin{verbatim}
s=%s;
num=2500;den=s*(s+5)*(s+50);
sys=syslin('c',num/den)
bode(sys)
show_margins(sys,'bode')
\end{verbatim}
 
#### 3.4.3 Diagrama de Bode de sistemas con retardos puros
El retardo añade una fase de $-\omega T_d$ radianes en la curva de fase sin afectar la curva de magnitud.
	
Por lo tanto, el máximo retardo que puede añadirse se calcula con:

\begin{equation}
\omega T_d=PM  \frac{\pi}{180^o}
\end{equation}
	
 
#### 3.4.4 Aproximaciones para sistemas con retardo

Para sistemas que incluyan el término $T_ds$, éste puede sustituirse mediante los siguientes polinomios:

Serie de Mcluarin:
\begin{equation}
e ^ {-T_d s} \approx \frac{1}{1+T_d s+ T_d ^ 2 s ^2  / 2}
\end{equation}

Serie de Padé:
\begin{equation}
e ^ {-T_d s} \approx \frac{1 - T_d s/2}{1 + T_d s/2}
\end{equation}
